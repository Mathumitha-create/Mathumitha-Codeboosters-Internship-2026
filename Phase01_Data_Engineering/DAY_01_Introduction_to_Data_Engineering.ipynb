```python
cs_students=df[df['department']=='Computer Science']
print("number of Computer Science Students:{len{cs_students}")
print("\nComputer Science Students:")
print(cs_students[['name','math_score','programming_score','attendance_percentage']])
```

```python
high_attendance=df[df['attendance_percentage']>90]
print("Students with attendance abouve 90%:{len(high_attendance)}")
print(high_attendance[['name','department','attendance_percentage']])
```

```python
df['total_Score']=df['math_score']+df['english_score']+df['science_score']+df['programming_score']
top_student=df.loc[df['total_Score'].idxmax()]
#idmax finds the index (row number) of the maximum value in total score
#df.loc selects a row by its index
print("Top Student:")
print(f"Name:{top_student['name']}")
print(f"Department:{top_student['department']}")
print(f"Total Score:{top_student['total_Score']}")
print(f"Attendance:{top_student["attendance_percentage"]}")
```

```python
dept_avg=df.groupby('department')['total_Score'].mean().round(2)
print("Avergae Total Score By Department:")
print(dept_avg)
```

```python
df_sorted=df.sort_values(by='total_Score',ascending=False)
print("Top 10 students by total Score")
print(df_sorted[['name','department','total_Score','attendance_percentage']].head(10))
```
